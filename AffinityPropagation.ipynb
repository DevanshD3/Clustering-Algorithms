```python
from sklearn.datasets import load_files
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.cluster import AffinityPropagation
```

```python

# for reproducibility
random_state = 0 

DATA_DIR = "/content/drive/MyDrive/Data/CV.csv"
df = pd.read_csv(DATA_DIR, low_memory= False)
df.head()
```

```python
vec = TfidfVectorizer(stop_words='english')
df['Modified Description'] = df['Modified Description'].fillna('')
vec_matrix = vec.fit_transform(df['Modified Description'])
features = vec.transform(df['Modified Description'])
```

```python
# reduce the features to 2D
pca = PCA(n_components=2, random_state=random_state)
reduced_features = pca.fit_transform(features.toarray())
```

```python
af = AffinityPropagation(preference=5)
clustering = af.fit(reduced_features)
```

```python
plt.scatter(reduced_features[:,0], reduced_features[:,1], c=clustering.labels_, alpha=0.7)
```
